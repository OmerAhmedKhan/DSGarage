```python
fig, ax = plt.subplots()
#ax.yaxis.set_major_formatter(formatter)
x = np.arange(4)
y = [100, 30.7, 19.4, 11.3]
plt.ylabel('%')
plt.title('Hit rate for biz vs private')
for i, v in enumerate(y):
    ax.text(i-0.25, v - 7, str(v)+"%", color='white', fontweight='bold', fontsize=14)

plt.bar(x, y,color='blue')
plt.xticks(x, ('Searches', 'Total Hits', 'Business', 'Private'))
plt.show()
```
